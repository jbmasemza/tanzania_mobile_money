```python
# Loading in required libraries/modules:
from IPython.display import display, HTML
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline




```

```python
#Reading the data set
financials = pd.read_csv("training.csv")
financials.head()
```

```python

#checking missing data 
missing_values = financials.isnull().sum().sum()
print('Number of missing values:',missing_values)
```

```python
#renaming columns name
financials = financials.rename(columns={
    'Q1':'age',
    'Q2':'gender',
    'Q3':'marital_status',
    'Q4':'highest_level_education ',
    'Q5':'plot_ownership',
    'Q6':'land_ownership',
    'Q7':'mobile_phone',
    'Q8_1':'salaries_wages',
    'Q8_2':'trading',
    'Q8_3':'service',
    'Q8_4':'piece_job',
    'Q8_5':'rental_income',
    'Q8_6':'investments',
    'Q8_7':'pension',
    'Q8_8':'grant',
    'Q8_9':'dependent',
    'Q8_10':'paid_expences',
    'Q8_11':'other1'})
```

```python
#Get basic descriptive statistics for the dataset.
financials.describe()
```

```python
# Filtering the original dataset in order to access the desired columns for the various financial services accessed
non = financials[(financials.mobile_money_classification == 0) | (financials.mobile_money_classification == 1)]

mobile = financials[(financials.mobile_money_classification == 2)]

both = financials[(financials.mobile_money_classification == 3)]
```

```python
# Age vs Financial services plots
age_non = non.age
age_mobile = mobile.age
age_both = both.age

plt.hist(age_non, color = ['red'])
plt.xlabel('Age distribution')
plt.ylabel('Number of people')
plt.title('The age distribution of people who do not use mobile money')
plt.show()

plt.hist(age_mobile, color = ['blue'])
plt.xlabel('Age distribution')
plt.ylabel('Number of people')
plt.title('The age distribution of people who only use mobile money')
plt.show()

plt.hist(age_both, color = ['green'])
plt.xlabel('Ages')
plt.ylabel('Number of people')
plt.title('The age distribution of people who use mobile money and one other financial service')
plt.show()
```

```python


# Relationship status
married = financials[financials.marital_status == 1]
divorced = financials[financials.marital_status == 2]
widowed = financials[financials.marital_status == 3]
single = financials[financials.marital_status == 4]

# status on mobile money classification
married_without = len(married[married.mobile_money_classification == 1])
married_with = len(married[married.mobile_money_classification == 2])
married_both = len(married[married.mobile_money_classification == 3])

divorced_without = len(divorced[divorced.mobile_money_classification == 1])
divorced_with = len(divorced[divorced.mobile_money_classification == 2])
divorced_both = len(married[married.mobile_money_classification == 3])

widowed_without = len(widowed[widowed.mobile_money_classification == 1])
widowed_with = len(widowed[widowed.mobile_money_classification == 2])
widowed_both = len(married[married.mobile_money_classification == 3])

single_without = len(single[single.mobile_money_classification == 1])
single_with = len(single[single.mobile_money_classification == 2])
single_both = len(married[married.mobile_money_classification == 3])

# sum of people with or without mobile money
relationship_with = [married_with, divorced_with, widowed_with, single_with]
relationship_without= [married_without, divorced_without, widowed_without, single_without]
relationship_both = [married_both,divorced_both,widowed_both,single_both]
#plotting the graph for relationship status and people with or without mobile money
plt.figure(figsize=(10,10))
plt.bar(np.arange(4), relationship_with, label='Without mobile money',color='blue')
plt.bar(np.arange(4), relationship_without, bottom= relationship_with, label='With mobile money',color='red')
plt.bar(np.arange(4), relationship_both, bottom= relationship_without, label='With  both mobile money and other financial service',color='yellow')
plt.xticks(np.arange(4), ['Married', 'Divorced', 'Widowed', 'Single'])
plt.legend(loc='upper right')
plt.title('Relationship Status vs Financials')

```

```python
# Type of income
salaries = financials[financials.salaries_wages == 1]
no_salaries = financials[financials.salaries_wages == 0]

trades = financials[financials.trading == 1]
no_trades = financials[financials.trading == 0]

services = financials[financials.service == 1]
no_services = financials[financials.service  == 0]

piece_work = financials[financials.piece_job == 1]
no_piece_work = financials[financials.piece_job == 0]

rent_income = financials[financials.rental_income == 1]
no_rent_income = financials[financials.rental_income == 0]

interest =financials[financials.investments == 1]
no_interest = financials[financials.investments == 0]

pensions = financials[financials.pension == 1]
no_pensions = financials[financials.pension == 0]

social_welfare = financials[financials.grant == 1]
no_social_welfare = financials[financials.grant == 0]

rely_on_someone = financials[financials.dependent == 1]
dont_rely_on_someone = financials[financials.dependent == 0]

dont_get_money = financials[financials.paid_expences == 1]
get_money = financials[financials.paid_expences == 0]

other = financials[financials.other1 == 1]
no_other = financials[financials.other1 == 0]

#summation
salaries = len(salaries[salaries.mobile_money == 1])
no_salaries = len(no_salaries[no_salaries.mobile_money == 0])
#both_salaries = len(salaries[salaries.mobile_money_classification == 3])

trades = len(trades[trades.mobile_money == 1])
no_trades = len(no_trades[no_trades.mobile_money == 0])
#both_trades = len(trades[trades.mobile_money_classification == 3])

services = len(services[services.mobile_money == 1])
no_services = len(no_services[no_services.mobile_money == 0])
#both_services = len(services[services.mobile_money_classification == 3])

piece_work = len(piece_work[piece_work.mobile_money == 1])
no_piece_work = len(no_piece_work[no_piece_work.mobile_money == 0])
#both_piece_work = len(piece_work[piece_work.mobile_money_classification == 3])

rent_income = len(rent_income[rent_income.mobile_money == 1])
no_rent_income = len(no_rent_income[no_rent_income.mobile_money== 0])
#both_rent_income = len(rent_income[rent_income.mobile_money_classification == 3])

interest = len(interest[interest.mobile_money == 1])
no_interest = len(no_interest[no_interest.mobile_money == 0])
#both_interest = len(interest[interest.mobile_money_classification == 3])

pensions = len(pensions[pensions.mobile_money == 1])
no_pensions = len(no_pensions[no_pensions.mobile_money == 0])
#both_pensions = len(pensions[pensions.mobile_money_classification == 3])

social_welfare = len(social_welfare[social_welfare.mobile_money == 1])
no_social_welfare = len(no_social_welfare[no_social_welfare.mobile_money == 0])
#both_social_welfare = len(social_welfare[social_welfare.mobile_money_classification == 3])

rely_on_someone = len(rely_on_someone[rely_on_someone.mobile_money == 1])
dont_rely_on_someone = len(dont_rely_on_someone[dont_rely_on_someone.mobile_money == 0])
#both_rely_on_someone = len(rely_someone[rely_someone.mobile_money_classification == 3])

dont_get_money = len(dont_get_money[dont_get_money.mobile_money == 0])
get_money = len(get_money[get_money.mobile_money == 1])
#both_get_money = len(get_money[get_money.mobile_money_classification == 3])

other = len(other[other.mobile_money == 1])
no_other = len(no_other[no_other.mobile_money == 0])
#both_other = len(other[other.mobile_money_classification == 3])

mobile_money = [salaries, trades, services, piece_work, rent_income, interest, pensions, 
      social_welfare, rely_on_someone, dont_get_money, other]
no_mobile_money = [no_salaries, no_trades, no_services, no_piece_work, no_rent_income, no_interest, no_pensions, 
         no_social_welfare, dont_rely_on_someone, get_money, no_other]
#both_mobile_money = [both_salaries, both_trades, both_services, both_piece_work, both_rent_income, both_interest, both_pensions, 
      #both_social_welfare, both_rely_on_someone, both_get_money, both_other]
```

```python

# Type of income
plt.figure(figsize=(15,10))
plt.barh(np.arange(11), mobile_money, label='With Mobile Money',color='blue')
plt.barh(np.arange(11), no_mobile_money, left= mobile_money, label='Without Mobile Money',color='red')
plt.yticks(np.arange(11), ['salaries', 'trades', 'services', 'piece_work', 'rent_income', 'interest', 'pensions', 
      'social_welfare', 'rely_on_someone', 'dont_get_money', 'other'])
plt.legend(loc='upper right')
plt.title('Type of Income and Mobile Money')
plt.show()
```

```python
# Land ownership
has_land = financials[financials.land_ownership == 1]
no_land = financials[financials.land_ownership == 2]

# Land ownership and Mobile Money
has_land_has_mobile_money = len(has_land[has_land.mobile_money == 1])
no_land_has_mobile_money= len(no_land[no_land.mobile_money == 1]) 

has_land_no_mobile_money = len(has_land[has_land.mobile_money == 0])
no_land_no_mobile_money = len(no_land[no_land.mobile_money == 0])
total = has_land_has_mobile_money + no_land_has_mobile_money + has_land_no_mobile_money + no_land_no_mobile_money

# Plotting Land ownership and Mobile Money
# Calculate piechart sizes
lmm = has_land_has_mobile_money / total * 100
nlmm = no_land_has_mobile_money / total * 100
lnmm = has_land_no_mobile_money / total * 100
nlnmm = no_land_no_mobile_money / total * 100

# Define the piechart
plt.figure(figsize=(10,5))
labels = ['Has land and Mobile Money','No land and Mobile Money','Has land and no Mobile Money','No land and no Mobile Money' ]
sizes = [lmm, nlmm, llmm,nnlwmm]
colors = ['red', 'blue', 'yellow', 'green']


# Plot the pie chart
plt.pie(sizes, explode=None, colors=colors, labels=labels, autopct='%1.1f%%',  startangle=140)
plt.title('Land Ownership')
plt.show()
```
